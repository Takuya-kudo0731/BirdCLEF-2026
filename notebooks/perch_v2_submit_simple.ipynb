# BirdCLEF+ 2026 — Submission (copy from training output)

Training notebook の Output から submission.csv をコピーして提出するだけのノートブック。

## Setup
1. Training notebook (`notebook17d3707d66`) の Output を Input として追加
2. Internet: OFF
3. Accelerator: None (CPU)
4. Submit to competition

```python
import os
import glob
import shutil
import pandas as pd

# Find submission.csv from training notebook output
candidates = glob.glob('/kaggle/input/**/submission.csv', recursive=True)
print(f'Found submission files: {candidates}')

assert candidates, (
    'submission.csv not found. '
    'Add the training notebook output as Input.'
)

src = candidates[0]
dst = '/kaggle/working/submission.csv'
shutil.copy(src, dst)
print(f'Copied: {src} -> {dst}')

sub = pd.read_csv(dst)
print(f'Rows: {len(sub)}')
print(f'Columns: {len(sub.columns)}')
print(f'Mean: {sub.drop("row_id", axis=1).values.mean():.6f}')
sub.head(3)
```
